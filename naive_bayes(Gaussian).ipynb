```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing

data = pd.read_csv('/content/gdrive/My Drive/dataset/buy_cp.csv')

age = data.iloc[:,0].values
income = data.iloc[:,1].values
student = data.iloc[:,2].values
credit_rating = data.iloc[:,3].values
buys_computer = data.iloc[:,4].values

le = preprocessing.LabelEncoder()
age = le.fit_transform(age)
income = le.fit_transform(income)
student = le.fit_transform(student)
credit_rating = le.fit_transform(credit_rating)
label = le.fit_transform(buys_computer)

features = list(zip(age, income, student, credit_rating))

model = GaussianNB()
model.fit(features, label)

print('Enter inputs')
a = int(input('Age : '))
i = int(input('Income : '))
s = int(input('Student : '))
cr = int(input('Credit Rating : '))
predicted = model.predict([[a, i, s, cr]])
print('\nPrediction :', predicted)
```
